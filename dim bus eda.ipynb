```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

bus = pd.read_csv("dim_bus.csv")
```

```python
num_cols = ['AvgEngineTemp', 'AvgRPM', 'TotalTrips']

for col in num_cols:
    plt.figure(figsize=(12,4))

    plt.subplot(1,2,1)
    sns.histplot(bus[col], kde=True)
    plt.title(f"Distribution of {col}")

    plt.subplot(1,2,2)
    sns.boxplot(x=bus[col])
    plt.title(f"Boxplot of {col}")

    plt.tight_layout()
    plt.show()
```

```python
cat_col = 'BusType'

plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
sns.countplot(x=cat_col, data=bus)
plt.title("Bus Type Countplot")
plt.xticks(rotation=45)

plt.subplot(1,2,2)
bus[cat_col].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Bus Type Percentage")
plt.ylabel("")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
sns.scatterplot(
    x='AvgEngineTemp',
    y='AvgRPM',
    data=bus
)
plt.title("Avg Engine Temp vs Avg RPM")

plt.subplot(1,2,2)
sns.boxplot(
    x='BusType',
    y='TotalTrips',
    data=bus
)
plt.title("Bus Type vs Total Trips")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
bus_num = bus[['AvgEngineTemp', 'AvgRPM', 'TotalTrips']]

corr = bus_num.corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
