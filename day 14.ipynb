```python
# Create fuel-related variables if they are missing

if 'FuelEfficiency' not in df.columns:
    df['FuelEfficiency'] = df['Distance'] / df['FuelConsumed']

if 'FuelPerKM' not in df.columns:
    df['FuelPerKM'] = df['FuelConsumed'] / df['Distance']

print(df[['FuelConsumed', 'FuelEfficiency', 'FuelPerKM']].head())
```
