```python
import pandas as pd

fact = pd.read_csv("sql fact file.csv")
```

```python
fact['FuelEfficiency'] = fact['Distance'] / fact['FuelConsumed']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr_data = fact[['IdleDuration', 'FuelConsumed']]

plt.figure(figsize=(6,4))
sns.heatmap(corr_data.corr(),
            annot=True,
            cmap='coolwarm',
            fmt='.2f')

plt.title('Idle Duration vs Fuel Consumed')
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=fact,
    x='AverageSpeed',
    y='FuelEfficiency'
)

plt.title('Average Speed vs Fuel Efficiency')
plt.xlabel('Average Speed (km/h)')
plt.ylabel('Fuel Efficiency (km/L)')
plt.show()
```

```python
pearson_corr = fact['AverageSpeed'].corr(
    fact['FuelEfficiency'],
    method='pearson'
)

print("Pearson Correlation:", pearson_corr)
```
