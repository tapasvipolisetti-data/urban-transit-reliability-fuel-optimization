```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

route = pd.read_csv("dim_route.csv")
```

```python
route.isnull().sum()
```

```python
route.duplicated().sum()
```

```python
route.info()
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)

sns.histplot(
    route['RouteDistanceKM'],
    kde=True
)

plt.title('Route Distance Distribution')

plt.subplot(1,2,2)

sns.boxplot(
    x=route['RouteDistanceKM']
)

plt.title('Route Distance Boxplot')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)

sns.countplot(
    x='RouteType',
    data=route
)

plt.title('Route Type Count')

plt.subplot(1,2,2)

route['RouteType'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.ylabel('')
plt.title('Route Type Percentage')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)

route['Source'].value_counts().head(10).plot(
    kind='bar'
)

plt.title('Top Source Locations')

plt.subplot(1,2,2)

route['Destination'].value_counts().head(10).plot(
    kind='bar'
)

plt.title('Top Destination Locations')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)

sns.boxplot(
    x='RouteType',
    y='RouteDistanceKM',
    data=route
)

```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)

sns.histplot(
    route['RouteDistanceKM'],
    kde=True
)

plt.title('Distance Distribution')

plt.subplot(1,2,2)

sns.kdeplot(
    route['RouteDistanceKM'],
    fill=True
)

plt.title('Distance Density')

plt.tight_layout()
plt.show()
```
